```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
temp2561=[20,35,28,15,35,19,20]
temp2560=[18,25,18,25,15,29,30]
temp2559=[30,25,38,19,15,41,18]


plt.plot(day,temp2561,color='red',linewidth=2,label='2561')
plt.plot(day,temp2560,color='blue',linewidth=2,label='2560')
plt.plot(day,temp2559,color='green',linewidth=2,label='2559')

plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather 7 Day !!!')
plt.legend()
plt.grid()
plt.show()
```
